```python
import pandas as pd
employees = pd.read_csv (r'C:\Users\firne\bip-python-unh\user_audit\data\employees.csv')
system_access = pd.read_csv (r'C:\Users\firne\bip-python-unh\user_audit\data\system_access.csv')
system_access.rename(columns={'id':'employee_number'}, inplace=True)
system_access.head()
```

```python
system_access.rename(columns={'id':'employee_number'}, inplace=True)
system_access['employee_number'] = system_access['employee_number'].apply(lambda x: '{0:0>06}'.format(x))
system_access['employee_number'] = 'EMP' + system_access['employee_number'].astype(str)
system_access.set_index('employee_number', inplace=True)
employees.set_index('employee_number', inplace=True)
df = employees.join(system_access, how= 'left')
system_access['access_level'] = system_access['access_level'].astype(str)
df.head()
```

```python
df.isnull().any()
nd = df[df.isnull().any(axis=1)]
df.loc[df['title'] == "Director of Sales", 'access_level'] = "Approval-Manager"
df.loc[df['title'] == "Account Manager", 'access_level'] = "Account Admin"
df.loc[df['title'] == "Sales Assistant", 'access_level'] = "Contract Entry"
df.head()
```

```python
df.to_csv(r'C:\Users\firne\bip-python-unh\user_audit\Final.csv')
```
